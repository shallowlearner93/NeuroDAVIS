```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import Datasets as DS
import Methods as Md
import DownstreamAnalysis as DA
from sklearn.metrics.pairwise import euclidean_distances
```

# Loading data

```python
X,y = DS.LoadData('Coil20')
```

```python
X.shape
```

```python
y = y.reshape((y.shape[0],))
```

# Applying T-SNE

```python
iteration = 10
res_knn_tsne = []
res_rfc_tsne = []
for i in range(iteration):
    print("Iteration {}".format(i+1))
    X_tsne = Md.UseMethod('TSNE', X)
    res_knn_tsne.append(DA.Knn(X_tsne, y, 5))
    res_rfc_tsne.append(DA.RFC(X_tsne, y))
```

```python
pd.DataFrame(res_knn_tsne,columns =['Acc','Pre','Rec','F1'])
```

```python
pd.DataFrame(res_rfc_tsne,columns =['Acc','Pre','Rec','F1'])
```

# Applying UMAP

```python
iteration = 10
res_knn_umap = []
res_rfc_umap = []
for i in range(iteration):
    print("Iteration {}".format(i+1))
    X_umap = Md.UseMethod('UMAP', X)
    res_knn_umap.append(DA.Knn(X_umap, y, 5))
    res_rfc_umap.append(DA.RFC(X_umap, y))
```
