# Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import Datasets as DS
import Methods as Md
import neuroDAVIS, Low = nd.NeuroDAVIS as ud
```

# Loading data

```python
dir = 'D:/Research/NeuroDAVIS/Results/'
X,y = DS.LoadData('FMnist')
```

```python
X.shape
```

```python
y = y.reshape((y.shape[0],))
```

```python
np.unique(y)
```

```python
c1=list(np.unique(y))
c2=['blue', 'green', 'orange', 'red', 'yellow','pink', 'purple', 'aquamarine', 'gray', 'darkgreen','skyblue', 'lime', 'royalblue' , 'cyan', 'olive',
    'tomato', 'saddlebrown', 'sienna', 'navy', 'violet', 'crimson', 'greenyellow','coral','burlywood','slategrey']
cols = map(lambda x: c2[c1.index(x)], y)
cs = [i for i in list(cols)]
```

```python
from sklearn.decomposition import PCA
X_pca = PCA(n_components=50).fit_transform(X)
```

# Applying T-SNE

```python
X_tsne = Md.UseMethod('TSNE', X)
```

# Plotting the results

```python
plt.figure(figsize=(12,8))
plt.scatter(X_tsne[:,0],X_tsne[:,1],c=cs,s=1)
plt.xlabel("T-SNE1", fontsize=16)
plt.ylabel("T-SNE2", fontsize=16)
plt.savefig(dir+'FMnist_tSNE_60k.eps', dpi = 1200, format = 'eps')
```

# Applying PCA + T-SNE

```python
X_pca_tsne = Md.UseMethod('TSNE', X_pca)
```

# Plotting the results

```python
plt.figure(figsize=(12,8))
plt.scatter(X_pca_tsne[:,0],X_pca_tsne[:,1],c=cs,s=1)
plt.xlabel("T-SNE1", fontsize=16)
plt.ylabel("T-SNE2", fontsize=16)
plt.savefig(dir+'FMnist_pca_tSNE_60k.eps', dpi = 1200, format = 'eps')
```

# Applying NeuroDAVIS

```python
data = np.eye(X.shape[0])
neuroDAVIS, Low = nd.NeuroDAVIS(data, X, dim=2, lambda_act=0, lambda_weight=0.0001, num_neuron=[500,1500], bs=512, epoch=1500, sd=1, verbose=1)
X_embedding = Low.predict(data)
```

# Plotting the results

```python
plt.figure(figsize=(12,8))
plt.scatter(X_embedding[:,0],X_embedding[:,1],c=cs,s=1)
plt.xlabel("NeuroDAVIS1", fontsize=16)
plt.ylabel("NeuroDAVIS2", fontsize=16)
plt.savefig(dir+'FMnist_NeuroDAVIS_60k.eps', dpi = 1200, format = 'eps')
```
