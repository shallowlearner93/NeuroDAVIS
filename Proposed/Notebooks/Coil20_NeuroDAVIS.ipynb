# Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import Datasets as DS
import NeuroDAVIS as nd
import DownstreamAnalysis as DA
from sklearn.metrics.pairwise import euclidean_distances
```

# Loading data

```python
dir = 'D:/Research/NeuroDAVIS/Results/HD/NeuroDAVIS/'
X,y = DS.LoadData('Coil20')
```

```python
c1=list(np.unique(y))
c2=['blue', 'green', 'orange', 'red', 'yellow','pink', 'purple', 'aquamarine', 'gray', 'darkgreen','skyblue', 'lime', 'royalblue' , 'cyan', 'olive',
    'tomato', 'saddlebrown', 'sienna', 'navy', 'violet','crimson', 'greenyellow','coral','burlywood','slategrey']
cols = map(lambda x: c2[c1.index(x)], y)
cs = [i for i in list(cols)]
```

# Applying NeuroDAVIS

```python
data = np.eye(X.shape[0])
neuroDAVIS, Low = nd.NeuroDAVIS(data, X, dim=2, lambda_act=0, lambda_weight=0, num_neuron=[100,400], bs=256, epoch=1000, sd=1, verbose=1)
X_embedding = Low.predict(data)
```

# Plotting the results

NeuroDAVIS

```python
plt.figure(figsize=(12,8))
plt.scatter(X_embedding[:,0],X_embedding[:,1],c=cs,s=10)
plt.xlabel("NeuroDAVIS1", fontsize=15)
plt.ylabel("NeuroDAVIS2", fontsize=15)
plt.savefig(dir+'Coil20_NeuroDAVIS.eps', dpi = 1200, format = 'eps')
```

# Spearman correlation

```python
iteration = 10
score = []
dist_orig = np.square(euclidean_distances(X, X)).flatten()
for i in range(iteration):
    print('iteration {}'.format(i+1))
    neuroDAVIS, Low = nd.NeuroDAVIS(data, X, dim=2, lambda_act=0, lambda_weight=0, num_neuron=[100,400], bs=256, epoch=1000, sd=0, verbose=0)
    X_embedding = Low.predict(data)
    dist_method = np.square(euclidean_distances(X_embedding, X_embedding)).flatten()
    score.append(DA.correlation(dist_orig, dist_method))
```
